```python
import pandas as pd
data = pd.read_csv("drive/My Drive/Colab Notebooks/Datasets/mushrooms.csv")
dataset = data
dataset
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

for col in dataset.columns:
    dataset[col] = labelencoder.fit_transform(dataset[col])

print(dataset)
```

```python
#split independent variable

dataset
X = dataset.iloc[:, 1:22].values
X
```

```python
#Split dependent variable
y = dataset.iloc[:, -1].values
y
```

```python
#Split data into Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
model = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
AdaBoost = AdaBoostClassifier(estimator=model, n_estimators=400, learning_rate=1, random_state=0)
boostmodel = AdaBoost.fit(X_train, y_train)

```

```python
y_pred = boostmodel.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
Accurancy = (cm[0][0]+cm[1][1])/(cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1])
Accurancy
```

```python
from sklearn import metrics
prediction = metrics.accuracy_score(y_test, y_pred)
print("Accuracy", prediction *100, '%')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
