```python
#An example of the classes in python:
class Person:
    institution ="MSIS"
    def __init__(self, name, age,location):
        self.name = name
        self.age = age
        self.location =  location

    def set_name(self, name):
        self.name = name
    def get_name(self):
        return self.name
    def set_age(self, age):
        self.age = age
    def get_age(self):
        return self.age
    def set_location(self, location):
        self.location = location
    def get_location(self):
        return self.location
```

```python

initialize_name = Person("John", 36,'manipal')
get_name = initialize_name.get_name()
print(get_name)
set_name = initialize_name.set_name("Sheethal Doe")
get_name = initialize_name.get_name()
print(get_name)
get_loacation = initialize_name.get_location()
print(get_loacation)
```
