```python
#split independent variable

dataset
X = dataset.iloc[:, 1:4].values
X
```

```python
#Split dependent variable
y = dataset.iloc[:, -1].values
y
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
X[:, 0] = labelencoder_X.fit_transform(X[:, 0])
print(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Different test sizes to check
test_sizes = [0.2, 0.25, 0.3]
```

```python
print("\n--- Decision Tree Confusion Matrices ---")
dt_conf_matrices = []
for ts in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ts, random_state=0)

    dt = DecisionTreeClassifier(criterion='gini', random_state=0)
    dt.fit(X_train, y_train)
    y_pred_dt = dt.predict(X_test)

    cm = confusion_matrix(y_test, y_pred_dt)
    dt_conf_matrices.append(cm)

    print(f"Test size = {ts}:\n{cm}")
```

```python
print("\n--- Logistic Regression Confusion Matrices ---")
log_conf_matrices = []
for ts in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ts, random_state=0)

    sc_X = StandardScaler()
    X_train = sc_X.fit_transform(X_train)
    X_test = sc_X.transform(X_test)

    log = LogisticRegression(random_state=0)
    log.fit(X_train, y_train)
    y_pred_log = log.predict(X_test)

    cm = confusion_matrix(y_test, y_pred_log)
    log_conf_matrices.append(cm)

    print(f"Test size = {ts}:\n{cm}")
```

```python
print("\n--- Linear Regression Confusion Matrices ---")
lin_conf_matrices = []
for ts in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ts, random_state=0)

    sc_X = StandardScaler()
    X_train = sc_X.fit_transform(X_train)
    X_test = sc_X.transform(X_test)

    lin = LinearRegression()
    lin.fit(X_train, y_train)
    y_pred_lin = lin.predict(X_test)

    # Convert to 0 or 1
    y_pred_lin_class = [1 if pred >= 0.5 else 0 for pred in y_pred_lin]

    cm = confusion_matrix(y_test, y_pred_lin_class)
    lin_conf_matrices.append(cm)

    print(f"Test size = {ts}:\n{cm}")
```

```python
print("\n--- Decision Tree Accuracy ---")
for ts, cm in zip(test_sizes, dt_conf_matrices):
    acc = (cm[0,0] + cm[1,1]) / cm.sum()   # Accuracy = (TP+TN)/Total
    print(f"Test size = {ts}, Accuracy = {acc}")
```

```python
print("\n--- Logistic Regression Accuracy ---")
for ts, cm in zip(test_sizes, log_conf_matrices):
    acc = (cm[0,0] + cm[1,1]) / cm.sum()
    print(f"Test size = {ts}, Accuracy = {acc}")
```
