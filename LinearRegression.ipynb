```python
#Linear regression example
import numpy as np
import pandas as pd
import statsmodels.api as sm

#create dataset
data = {
    'Hours_Studied': [2, 4, 6, 8, 10],
    'Classes_Attended': [3, 4, 5, 6, 8],
    'Final_Score': [65, 70, 75, 85, 95]
}
df = pd.DataFrame(data)

#Define independent variables (X) and dependent variable (Y)
X = df[['Hours_Studied', 'Classes_Attended']]
Y = df['Final_Score']

#Add a constant to the independent variables
X = sm.add_constant(X)
X

```

```python
#Fit the linear regression model
model = sm.OLS(Y, X).fit()

#print the summary
print(model.summary())
```

```python
#predict final score for a new student who studied 7 hrs and attended 5 classes
new_data = pd.DataFrame({'const': [1], 'Hours_Studied': [7], 'Classes_Attended': [5]})
predicted_score = model.predict(new_data)[0]
print(f"Predicted final score for a student who studied 7 hrs and attended 5 classes: {predicted_score}")
```
