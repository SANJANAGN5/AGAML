```python
import pandas as pd
data = pd.read_csv("drive/My Drive/Colab Notebooks/Datasets/DataOLD.csv")
dataset = data
dataset
```

```python
#split independent variable

dataset
X = dataset.iloc[:, :-1].values
X
```

```python
#Split dependent variable
y = dataset.iloc[:, 3].values
y
```

```python
#Taking care of missing values

#fill the "Nan" with mean of all the values
#assignment - most occuring number to fit in the "Nan"

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(X[:, 1:3]) #imputer is a class
X[:, 1:3] = imputer.transform(X[:, 1:3])
X
```

```python
#Encoding the Categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
X[:, 0] = labelencoder_X.fit_transform(X[:, 0])
print(X)
```

```python
#Encoding the Independent Variable

from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("Country", OneHotEncoder(),[0])], remainder="passthrough")
X = ct.fit_transform(X)
X.astype(int)
```

```python
#Encoding the Dependent Variable
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
y
```

```python
#Split data into Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```
